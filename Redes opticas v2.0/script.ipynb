```python
import gymnasium as gym
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from IPython.display import display
from stable_baselines3 import PPO
from stable_baselines3.common.vec_env import DummyVecEnv
from stable_baselines3.common.utils import set_random_seed
from custom_env.redes_opticas_env import RedesOpticasEnv

def make_env(num_ont, Vt=10e6, rank=0, seed=0):
    """
    Utility function for multiprocessed env.
    
    :param num_ont: Número de ONTs en la red
    :param Vt: Capacidad de tráfico
    :return: un entorno con configuración específica
    """
    def _init():
        env = RedesOpticasEnv(render_mode=None, num_ont=num_ont, Vt=Vt)
        return env
    return _init


if __name__ == "__main__":
    env_id = 'RedesOpticasEnv-v0'  # Hay que asegurarse de que este ID coincida con el registrado
    num_test = 10  #Ponemos el numero de test que necesitamos para que el algoritmo de aprendizaje aprenda.
    seed = np.random.randint(0, 10)  #Ponemos seeds aleatorias
    num_envs = 4  # Número de entornos paralelos
    #EL NUMERO DE ONTS PUEDE SER DEFINIDO DESDE ESTA VARIABLE(POSIBLE INPUT)
    num_ont=5
    #Establecemos el Vt
    Vt=10e6
    #Transmision de cada ciclo
    T=0.002
    # Vector con los entornos establecidos
    vec_env = DummyVecEnv([make_env(num_ont, Vt, rank=i, seed=42) for i in range(num_envs)])
    # Definimos el modelo, en nuestro caso el PPO con MlpPolicy
    model = PPO("MlpPolicy", vec_env, verbose=1, n_steps=2048, learning_rate=0.00025, gamma=0.99, gae_lambda=0.95)
    model.learn(total_timesteps=10000)

    # Fase de pruebas

    #Establecemos un unico episodio a investigar
    num_test_episodes = 1  # Número de episodios de prueba

    # Lista para guardar la información de cada episodio
    episode_info = []  

    # Lista de en cada ont guardar el valor de su capacidad
    list_ont = [[] for _ in range(num_ont)]

    # Guardar los estados de ON y OFF del estado de pareto
    estados_on_off_recolectados = [[] for _ in range(num_ont)]

    # Ejecutamos cada episodio.

    for episode in range(num_test_episodes):
        obs = vec_env.reset()  # Resetea el entorno al estado inicial
        done = np.array([False]*num_envs)  # Inicializa 'done' para todos los entornos

        while not done.all():
            action, _states = model.predict(obs, deterministic=True)  # Usamos el modelo para predecir la acción
            obs, rewards, dones, info = vec_env.step(action)

            # Guardamos la informacion del episodio.
            episode_info.append(info)
            for i in range(len(info)):  # Itera sobre cada sub-entorno
                
                suma=0
                for j in range(num_ont):
                    list_ont[j].append(info[i]['band_onts'][j])
                    estados_on_off_recolectados[j].append(info[i]['on_off_states'][j])
                    suma+=info[i]['band_onts'][j]
                
                #En caso de que la suma supere el limite establecido, el Vt*T decimos de que no debe superar el valor establecido y vemos cuanta es la desviacion.
                if suma>Vt*T:
                        dif=Vt*T-suma
                        print(f'SALE MAL en el ciclo {i} y con una diferencia del tope es de {dif}')
            done |= dones  # Actualiza 'done' para todos los entornos

    #print(episode_info[0])
    
    #print(list_ont)

    #print(estados_on_off_recolectados)
    
    #Mostrar informacion de los datos

    for i in range(num_ont):
        
        #Grafica del trafico de pareto de las redes
        plt.figure(figsize=(12, 6))
        plt.ylim(0,23000)
        plt.xlim(0,100)
        plt.plot(list_ont[i], label=f'Grafica de la ONT {i+1}')
        plt.title(f'Grafica de la ONT {i+1}')
        plt.show()


        #Print de los estados de OFF y ON.
        print(estados_on_off_recolectados[i])
        
        # Grafica de el trafico de el cambio de estado de pareto.
        plt.figure(figsize=(12, 6))
        plt.ylim(0,2)
        plt.xlim(0,100)
        plt.plot(estados_on_off_recolectados[i], label=f'Grafica de la ONT: {i+1} ON_OFF')
        plt.title(f'Grafica de la ONT {i+1} ON_OFF')
        plt.show()


    
```
